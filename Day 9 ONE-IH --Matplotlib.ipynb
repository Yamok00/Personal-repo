```python
# importing the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [5,10,15]
```

```python
# Exploring some matplotlib functions
plt.plot(x,y, label='Amount Sold', color='green', marker='*')

plt.title('First graph', fontdict={'color':'blue',
                                     'fontsize':15,
                                     'fontname':'Algerian'})
plt.xlabel('X axis', fontdict={'color':'green',
                              'fontsize':'20'})
plt.ylabel('Y axis', fontdict={'color':'blue',
                              'fontsize':20})

# Placing a lagend on the label
# bbox_to_anchor to move the legend
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1))

# Setting the scale of the graph
plt.xticks([0.5,1,1.5,2,2.5,3])
plt.yticks([2.5,5,7.5,10,12.5,15])

plt.show()
```

```python
a = [2,4,6,8,10]
b = [5,10,15,20,25]
# Using shorthand notation
#Note the shorthand notation for the coloour,marker and linestyle ('b*-')
plt.plot(a,b, 'b*-.',label='Enjoyment')

plt.title('Second Graph!', fontdict={'color':'red',
                                    'fontsize':10})

plt.legend(title='Aiye', bbox_to_anchor=(1, 1))

plt.show()
```

```python
# Creating data using numpy
x1 = np.arange(0,4.5,0.5)
x2 = x1**2
```

```python
# Plotting multiple graphs
plt.plot(x,y, label='Amount Sold', color='green', marker='*')
plt.plot(a,b, 'b*-.', label='Enjoyment')
plt.plot(x1,x2, 'r*-.', label='Third')

plt.legend(title='Title', bbox_to_anchor=(1,1))
plt.show()
```

```python
# a sketch to show sudden change in data
plt.plot(x1[:6], x2[:6], 'g*-', label='A')
plt.plot(x1[5:], x2[5:], 'b^--', label='B')


plt.legend(title='Category', bbox_to_anchor=(1,1))
# Saving the graph
plt.savefig('Third Graph.png')

plt.show()
```

### Barchart with matplotlib

```python
labels = ['Jan', 'Feb', 'Mar']
values = [500, 1000, 1500]
```

```python
plt.figure(figsize=(4,3))
bars = plt.bar(labels,values, color='red')
plt.title('First Barchart', color = 'green', fontdict ={'fontname':'Algerian'})
plt.xticks(rotation=45)

bars[0].set_hatch('o')
bars[1].set_hatch('*')
bars[2].set_hatch('/')

plt.show()
```
